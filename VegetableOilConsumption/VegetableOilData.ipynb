### Vegetable Oil Consumption Per Capita

```python
df = pd.read_excel('VegetableOilConsumptionPerCapita.xlsx')
```

```python
#Visualizing Data
df.head()
```

```python
#Analyzing shape of array (rows, columns)
df.shape
```

```python
#Resetting the labels
df.set_axis([n for n in range(1959, 2015)], axis='columns', inplace=True)
```

```python
#Renaming Columns
df.rename(columns={1959 : 'Countries', 1960 : 'Units'}, inplace=True)
df.head()
```

```python
#Dropping Source Column
df.drop([2014], axis=1, inplace=True)
```

```python
#Dropping yearly Data row
df.drop([0], axis=0, inplace=True)
df.head()
```

```python
#Observing Data
df.info()
```

```python
#Checking There are unique values
df['Units'].unique()
```

```python
#Checking duplicates
df.duplicated(subset=['Units']).sum()
```

```python
#Checking duplicates
df.duplicated(subset=['Countries']).sum()
```

```python
df.head()
```

```python
#Cleaning locked values
data = df.drop(df.loc[:, 1961:2006].columns, axis=1).drop(df[[2012, 2013]],axis=1)
```
